```python
####Get a Pandas DataFrame for Daily Confirmed Cases in Nigeria. Columns are Date and Cases
import pandas as pd

# Load the global daily confirmed cases data from the GitHub URL
url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv"
df = pd.read_csv(url)

# Check the first few rows to understand the data structure
print(df.head())

# Filter the data for Nigeria (Assuming the country is in the 'Country/Region' column)
nigeria_df = df[df['Country/Region'] == 'Nigeria']

# Drop unnecessary columns (keep only the Date columns and 'Confirmed' column)
nigeria_df = nigeria_df.drop(['Lat', 'Long', 'Country/Region'], axis=1)

# Transpose the data to make the dates the index
nigeria_df = nigeria_df.transpose()

# Rename the columns for clarity
nigeria_df.columns = ['Cases']

# Reset the index to make it a column
nigeria_df = nigeria_df.reset_index()

# Rename the columns to 'Date' and 'Cases'
nigeria_df.columns = ['Date', 'Cases']

# Ensure we remove any rows where the 'Date' column has non-date values
nigeria_df = nigeria_df[nigeria_df['Date'] != 'Province/State']

# Convert the 'Date' column to datetime format with an explicit format (MM/DD/YY)
nigeria_df['Date'] = pd.to_datetime(nigeria_df['Date'], format='%m/%d/%y', errors='coerce')

# Drop rows where 'Date' is NaT (Not a Time) due to parsing issues
nigeria_df = nigeria_df.dropna(subset=['Date'])

# Display the cleaned DataFrame
print(nigeria_df.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# URLs to the time series data for confirmed, recovered, and death cases
confirmed_url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv"
recovered_url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv"
deaths_url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv"

# Load the data from the URLs
confirmed_df = pd.read_csv(confirmed_url)
recovered_df = pd.read_csv(recovered_url)
deaths_df = pd.read_csv(deaths_url)

# Filter the data for Nigeria
nigeria_confirmed = confirmed_df[confirmed_df['Country/Region'] == 'Nigeria'].drop(['Country/Region', 'Province/State'], axis=1).transpose()
nigeria_recovered = recovered_df[recovered_df['Country/Region'] == 'Nigeria'].drop(['Country/Region', 'Province/State'], axis=1).transpose()
nigeria_deaths = deaths_df[deaths_df['Country/Region'] == 'Nigeria'].drop(['Country/Region', 'Province/State'], axis=1).transpose()

# Remove metadata rows from the dataframe (e.g., "Lat", "Long")
nigeria_confirmed = nigeria_confirmed[1:]  # Skip the first row containing metadata
nigeria_recovered = nigeria_recovered[1:]
nigeria_deaths = nigeria_deaths[1:]

# Explicitly specify the format for the dates
date_format = "%m/%d/%y"  # Date format can be adjusted depending on the dataset's format
nigeria_confirmed.index = pd.to_datetime(nigeria_confirmed.index, format=date_format, errors='coerce')
nigeria_recovered.index = pd.to_datetime(nigeria_recovered.index, format=date_format, errors='coerce')
nigeria_deaths.index = pd.to_datetime(nigeria_deaths.index, format=date_format, errors='coerce')

# Calculate daily cases by subtracting the previous day's total
daily_confirmed = nigeria_confirmed.diff().fillna(0)
daily_recovered = nigeria_recovered.diff().fillna(0)
daily_deaths = nigeria_deaths.diff().fillna(0)

# Plot the total daily confirmed, recovered, and death cases
plt.figure(figsize=(12, 6))

# Plot daily confirmed cases
plt.plot(daily_confirmed.index, daily_confirmed.values, label='Daily Confirmed Cases', color='blue')

# Plot daily recovered cases
plt.plot(daily_recovered.index, daily_recovered.values, label='Daily Recovered Cases', color='green')

# Plot daily death cases
plt.plot(daily_deaths.index, daily_deaths.values, label='Daily Deaths Cases', color='red')

# Add labels and title
plt.xlabel('Date')
plt.ylabel('Number of Cases')
plt.title('Total Daily Confirmed, Recovered, and Death Cases in Nigeria')
plt.legend()

# Format the x-axis labels
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#### Total Daily Confirmed, Recovered, and Death Cases in Nigeria
"Observation:
"The plot displays significant fluctuations in the confirmed, recovered, and death cases over time.
Peaks and drops represent periods of intense infection waves or reporting irregularities.
The green line (recovered cases) appears to dip unusually, potentially due to data entry errors or reporting inconsistencies."
```
